```python
import matplotlib.pyplot as plt

# Data for CPI basket
labels = ['Food & Beverages (45.86%)', 'Housing (10.07%)', 'Fuel & Light (6.84%)',
          'Clothing & Footwear (6.53%)', 'Pan, Tobacco & Intoxicants (2.38%)',
          'Miscellaneous (28.32%)']
sizes = [45.86, 10.07, 6.84, 6.53, 2.38, 28.32]

plt.figure(figsize=(8,8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('CPI Basket Breakdown (India)')
plt.show()
```
